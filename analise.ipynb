### Tratamento dos dados

```python
# Visualização dos dados de colunas do df_vendas
lista_colunas = ["Loja", "Cliente"]
display(df_vendas[lista_colunas])
```

```python
# Exluindo a coluna Data_Base do df_analise
df_analise = df_vendas.drop(columns=["Data_Base"])
display(df_analise)
```

```python
# Preenchimento dos valores vazios da coluna "Loja" com o texto "Online"
df_analise["Loja"] = df_analise["Loja"].fillna("Online")
display(df_analise)
```

```python
# Transformar o tipo de dado da coluna para datetime
df_analise["Data"] = pd.to_datetime(df_analise["Data"], format="%Y-%m-%d")
display(df_analise)
display(df_analise.info())
```

```python
# Padronização
df_analise["Loja"] = df_analise["Loja"].str.strip()
df_analise["Loja"] = df_analise["Loja"].str.title()
```

```python
# Removendo duplicatas
df_analise = df_analise.drop_duplicates(subset=["ID_Pedido", "Loja"])
```

### Feature Engineering

```python
# Criando nova colunas

# Faturamento
df_analise["Faturamento"] = df_analise["Qtd"] * df_analise["Preco_Unitario"]

# Forma de venda
df_analise["Forma_de_venda"] = np.where(df_analise["Loja"]=="Online", "Online", "Presencial")

# Região
display(df_analise["Loja"].unique())
```

```python
# Coluna Região

dic_regioes = {
  "São Paulo": "Sudeste",
  "Rio De Janeiro": "Sudeste",
  "Belo Horizonte": "Sudeste",
  "Salvador": "Nordeste",
  "Recife": "Nordeste",
  "Curitiba": "Sul",
  "Porto Alegre": "Sul",
  "Online": "Online"
}

df_analise["Região"] = df_analise["Loja"].map(lambda cidade: dic_regioes.get(cidade, "Indeterminado"))

display(df_analise["Região"])
```

### Análise

#### Filtrar

```python
i = 3

# .loc -> por indice da linha e nome da coluna

loja = df_analise.loc[i, "Loja"]
produto, cliente = df_analise.loc[i, ["Produto", "Cliente"]]

print(loja)
print(produto)
print(cliente)

# .iloc -> por posição

id_pedido = df_analise.iloc[0, 0]
print(id_pedido)
```

```python
# Exportar uma parte da base de dados

# Vendas das loja de Curitiba
df_vendas_cw = df_analise[df_analise["Loja"] == "Curitiba"]

# Exportando para arquivo csv
df_vendas_cw.to_csv("vendas_loja_curitiba.csv", index=False)

# Visualização do df_venda_cw
display(df_vendas_cw)
```

```python
# Anos que possuem vendas
vendas_anos = df_analise["Data"].dt.year.unique().tolist()
print(vendas_anos)

# Vendas de 2024

df_vendas_2024 = df_analise[df_analise["Data"].dt.year == 2024]
display(df_vendas_2024)
```

```python
# Dupla Condições

c1 = df_analise["Produto"] == "Cabo HDMI"
c2 = df_analise["Região"] == "Sul"
df_venda_HDMI_e_sul = df_analise[c1 & c2]

display(df_venda_HDMI_e_sul)
```

#### Agrupamento

```python
# Faturamento total de cada loja
faturamento_lojas = df_analise[["Loja", "Faturamento"]].groupby("Loja").sum()
faturamento_lojas = faturamento_lojas.sort_values(by="Faturamento", ascending=False)
faturamento_lojas = faturamento_lojas.reset_index()
faturamento_lojas = faturamento_lojas["Faturamento"].map("R$ {:,.2f}".format)
display(faturamento_lojas)
```

```python
# Média de faturamento das lojas
media_lojas = df_analise[["Loja", "Faturamento"]].groupby("Loja", as_index=False).mean()
media_lojas = media_lojas.sort_values(by="Faturamento", ascending=False)
display(media_lojas)
```

```python
# analises por agrupamento
# display(df_analise)

# ranking de faturamento por loja
analise_lojas = df_analise[["Loja", "Faturamento"]].groupby("Loja").sum()
analise_lojas = analise_lojas.sort_values(by="Faturamento", ascending=False)
analise_lojas = analise_lojas.reset_index()
analise_lojas["Faturamento"] = analise_lojas["Faturamento"].map("R${:,.2f}".format)
display(analise_lojas)

# ranking de produtos que mais venderam no online
df_vendas_online = df_analise[df_analise["Loja"]=="Online"]
analise_produtos_online = df_vendas_online[["Produto", "Qtd"]].groupby("Produto").sum()
analise_produtos_online = analise_produtos_online.sort_values(by="Qtd", ascending=False)
analise_produtos_online = analise_produtos_online.rename(columns={"Qtd": "Vendas Totais"}) # alterar nome de coluna
display(analise_produtos_online)

# analise de rankind por loja e por produto
# quais produtos venderam mais em cada uma das lojas
analise_produtos_em_lojas = df_analise[["Loja", "Produto", "Qtd"]].groupby(["Loja", "Produto"]).sum()

# quais lojas mais venderam os produtos
analise_lojas_em_produtos = df_analise[["Loja", "Produto", "Qtd"]].groupby(["Produto", "Loja"]).sum()

with pd.option_context("display.max_rows", None):
    display(analise_produtos_em_lojas)
    display(analise_lojas_em_produtos)
```

```python
display(df_analise.head())

# quais gerentes bateram a meta em janeiro de 2023
df_meta = df_analise[(df_analise["Data"].dt.year==2023) & (df_analise["Data"].dt.month==1)]

df_meta = df_meta[["Loja", "Faturamento"]].groupby("Loja", as_index=False).sum()

df_meta = df_meta.merge(df_gerentes, on="Loja", how="left")
df_meta["Bateu Meta"] = np.where(df_meta["Faturamento"] >= df_meta["Meta_Mensal"], "Sim", "Não")
display(df_meta)
```
